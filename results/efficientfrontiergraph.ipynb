```python
import numpy as np
import matplotlib.pyplot as plt

# Asset parameters (same example values as before)
mu = np.array([0.08, 0.12])            # Expected returns
sigma = np.array([0.15, 0.25])         # Standard deviations
corr = 0.3                             # Correlation coefficient

# Covariance matrix
cov = np.array([[sigma[0]**2, corr * sigma[0] * sigma[1]],
                [corr * sigma[0] * sigma[1], sigma[1]**2]])

# Generate efficient frontier
weights = np.linspace(0, 1, 200)
port_returns = weights * mu[0] + (1 - weights) * mu[1]
port_variances = (
    weights**2 * cov[0, 0]
    + 2 * weights * (1 - weights) * cov[0, 1]
    + (1 - weights)**2 * cov[1, 1]
)
port_risks = np.sqrt(port_variances)

# Find the MVP
min_idx = np.argmin(port_risks)
mvp_risk = port_risks[min_idx]
mvp_return = port_returns[min_idx]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(port_risks, port_returns, color='blue', label="Efficient Frontier")
plt.scatter(mvp_risk, mvp_return, color='blue', s=80, label="MVP")

# Labels
plt.xlabel("Risk (Std. Deviation)")
plt.ylabel("Expected Return")
#plt.title("")

# Remove axis tick labels
ax = plt.gca()
ax.tick_params(axis='both', which='both', labelbottom=False, labelleft=False)

plt.grid(True)
plt.legend()

# Save as PDF
file_path = 'efficient_frontier_graph.pdf'
plt.savefig(file_path, format='pdf')

plt.show()

file_path
```
